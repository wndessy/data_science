```python
%matplotlib inline
def show_stock_features(filename,timeperiodFrom,timeperiodTo):
    data=pd.read_csv(filename)
    df = pd.DataFrame(data)
    # Set the Date as Index
    df.index = pd.to_datetime(df.Date)

    # Select the range of data to be plotted
    filteredDF=df[['Open','High','Low','Close']][timeperiodFrom:timeperiodTo]
    
    busines_daily=filteredDF.resample('B').ffill()
    monthly=filteredDF.resample('M').mean().ffill()

    busines_daily.plot(figsize=(25, 6) , title='Daily data plot')
    monthly.plot(figsize=(25, 6), title='Monthly data plot')


show_stock_features(filename=ahpi,timeperiodFrom='2008-01',timeperiodTo='2011-12')
show_stock_features(filename=aker,timeperiodFrom='2014-01-24',timeperiodTo='2017-01-24')
```

```python
%matplotlib inline
import matplotlib.dates as mdates
from pandas.tseries.offsets import MonthOffset
import calendar

def show_multiple_stocks(stocknameslist,period_start,period_end):

#   initialize pyplot figure and create a subplot  
    fig = plt.figure(figsize=[25,10])
    ax = fig.add_subplot(111)
    ax.set_title('Share Close Values Plot')
    ax.set_xlabel('months of the year')
    ax.set_ylabel('normalized ‘Close’ values')
    
 #  Select Stock Closevalue for each stock and create a new dataframe
    newdf=pd.DataFrame()
    for item in stocknameslist:
        df= pd.DataFrame(pd.read_csv(item))
        df.index = pd.to_datetime(df.Date)
        #     select the desired range and fill in missing data
        close_value=df.loc[period_start:period_end].Close.ffill()
        #     normalize data
        normalized=((close_value-close_value.min())/(close_value.max()-close_value.min()))
#         *100#=====================================
        #     Merge  onto the main dataframe
        name=item.split('/')[1]
        newdf=newdf.join(normalized, how='outer' ,rsuffix='_'+name)
        #   Plot the current dataset close value
        line=ax.plot(normalized,lw=1,label=name)
    #     line=ax.plot(newdf,lw=1,label=name)    
    newdf.ffill(inplace=True)
    ax.legend(loc='best')
    # Preparing and creating pointers 
    octstackData=newdf['2016-9'].stack()
    ax.annotate(s='Highest in October',xy=(octstackData.idxmax()[0],octstackData.max()),
                arrowprops=dict(facecolor='black',arrowstyle='->'),
                horizontalalignment='left', verticalalignment='bottom', color='purple')
    febstackData=newdf['2016-2'].stack()    
    ax.annotate(s='Lowest in February',xy=(febstackData.idxmin()[0],febstackData.min()),
                arrowprops=dict(facecolor='black',arrowstyle='->'),
                horizontalalignment='left', verticalalignment='top', color='brown'
               )
    #Setting tick labels
    months=newdf.resample('MS').first().index.month.map(lambda x:calendar.month_name[x])
    ticks=newdf.resample('MS').first().index
    ticks=ax.set_xticks(ticks)
    labels=ax.set_xticklabels(months,rotation=30,fontsize='large')
    plt.show()

    return newdf

# result=show_multiple_stocks(stocknameslist=[aapl,aker,ahpi,amzn,msft,hmc,nflx,tho,tm],period_start='2012',period_end='2016')
result=show_multiple_stocks(stocknameslist=[tho,tm,hmc],period_start='2016.01.04',period_end='2016.12.30')

result
```

## 3.1 ARIMA
In this part, you can create ARIMA model from statsmodels.tsa.arima_model
library of Python to provide required forecasts. The procedure for doing
prediction using ARIMA is as follows:
1. Read the dataset for ‘aapl’ stock and select a subset of data containing
information from 2014 to 2016 (including 2016) .Fill in the missing dates.
2. Create a new data frame out of this subset, where ‘Date’column is set as index
column and ‘Open’ price is set as another column.
3. Divide the dataset into two sets: Trainset including daily data of the years 2014
and 2015 and test set containing data in 2016.
4. Define the model by calling ARIMA and passing in the p, d, and q parameters
with 3,1 and 0 values respectively. This sets the lag value to 3 for autoregression,
uses a difference order of 1 to make the time series stationary, and uses a moving
average model of 1.
5. To have a one-step forecast, for each sample in testset, create an ARIMA model
on the training data (Numpy values of ‘Open’ prices in train set) by calling the
‘fit’ function and setting the ‘disp’ argument value to zero. Then call the ‘forecast’
function of the fitted model and store its first returned argument as the predicted
value for the given test sample.
6. Store all predictions in a list and evaluate the prediction performance using
square root of ‘mean_squared_error’ metric from sklearn.metrics package. This
metric calculates the prediction error in terms of RMSE (Root Mean Squared
Error).
7. Plot the predictions against the real values in testset.

```python
from  statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
import math  
data=pd.read_csv(aapl)
df = pd.DataFrame(data)
# Set the Date as Index
df.Date=pd.to_datetime(df.Date)
# Select the range =============filll missing values
df.set_index('Date',inplace=True)
filteredf=df.loc['2014':'2016']
newdf=filteredf.Open
# dividing the dataset into test and training sets
train=newdf.loc['2014':'2015']
test=newdf.loc['2016']
# initializing training set and prediction values
training_set=train.values
predictions=[];
for i,v in test.items():
#     train and fit the model
    training_set=np.append(training_set,test[:i].values)
    model = ARIMA(training_set, order=(3,1,0))
    model_fit = model.fit(disp=0)
    forecast=model_fit.forecast()[0]
#     print (forecast)
#      store all the predictions of forecast
    predictions.append(forecast)
#     break # used for testing 

# Compute root mean square error
rmse = math.sqrt(mean_squared_error(test.values, predictions))
print('RMSE: ', rmse)
# # Plot data
fig1 = plt.figure(figsize=[25,10])
ax1 = fig1.add_subplot(111)
ax1.set_title('ARIMA Model Predictions plot')
ax1.set_xlabel('Tests')
ax1.set_ylabel('Predictions')
# ax1.plot(test.values, predictions)
ax1.plot(test.values,label='test')
ax1.plot(predictions,label='predictions')
ax1.legend(loc='best')
plt.show
 

```

## 3.2 Random Forest
To provide the predictions using the random forest, you should follow the
following steps:
1. Read the dataset for ‘aapl’ stock and select a subset of data containing
information from 2012 to 2016. Fill in the missing dates. Convert it to a new
dataset.
2. Set the ‘Date’ column as the index of the new dataset and add a new column
to the created subset: the ‘Open’ Price of next day. (This would be our target
variable, and this is made by shifting value of ‘Open’ price of current day by one)
3. Convert the data frame to a Numpy series and use its values to create two sets:
Trainset including daily data of the years 2014 and 2015 and test set containing
data in 2016.
4. Divide the trainset to train_X and train_y. Divide the testset to test_X and test_y
where train_X includes information about stock features and train_y is the ‘Open’
price value of next day (prediction target value). The same applies to testset.
5. From sklearn.ensemble import RandomForestRegressor
6. Build a random forest regressor model using default parameters.
7. Call ‘fit’ function of the created model using two arguments: Train_X and
Train_y.
8. Create predictions by calling ‘predict’ function of the fitted model using one
argument: Test_X .
9. Use RMSE metric error again to compute the prediction error rate (feed it with
two arguments: test_y and predictions).
10. Plot the predictions against real values.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import math  

data=pd.read_csv(aapl)
df = pd.DataFrame(data)
# Set the Date as Index
df.Date=pd.to_datetime(df.Date)
# Select the range =============filll missing values
df.set_index('Date',inplace=True)
filteredf=df.loc['2014':'2016']
# adding new collumn
filteredf['Open_price_value_of_next_day']=filteredf.Open.shift(1).bfill()
# # dividing the dataset into test and training sets
train_set=filteredf.loc['2014':'2015']
test_set=filteredf.loc['2016']

train_x=train_set[['High','Low','Open','Volume','Close']].values
train_y=train_set.Open_price_value_of_next_day.values
test_x=test_set[['High','Low','Open','Volume','Close']].values
test_y=test_set.Open_price_value_of_next_day.values

# initializing random forest
# regr = RandomForestRegressor(max_depth=2, random_state=0,n_estimators=100)
regr = RandomForestRegressor()
model=regr.fit(train_x, train_y)
predictions=model.predict(test_x)

# Compute root mean square error
rmse = math.sqrt(mean_squared_error(test_y,predictions))
print('RMSE: ', rmse)
# # Plot data
fig1 = plt.figure(figsize=[25,10])
ax1 = fig1.add_subplot(111)
ax1.set_title('Random Forest Model Predictions plot')
ax1.set_xlabel('Tests')
ax1.set_ylabel('Predictions')
ax1.plot(test_y,label='test')
ax1.plot(predictions,label='predictions')
ax1.legend(loc='best')
plt.show
 




```
